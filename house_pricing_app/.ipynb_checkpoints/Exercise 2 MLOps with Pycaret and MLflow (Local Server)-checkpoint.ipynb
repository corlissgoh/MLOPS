# MLOps con MLflow

#### Get data

```python
import mlflow

mlflow.set_tracking_uri("http://127.0.0.1:5000")

# load the dataset from pycaret
from pycaret.datasets import get_data
data = get_data('diamond')
```

#### Preprocessing

```python
# initialize setup
from pycaret.regression import *
s = setup(data, target = 'Price', 
          transform_target = True, 
          log_experiment = True, 
          #log_plots = True,
          experiment_name = 'diamond')
```

#### Model Training

```python
# compare all models
best = compare_models()
```

```python
# check the residuals of trained model
plot_model(best, plot = 'residuals')
```

```python
# check feature importance
plot_model(best, plot = 'feature')
```

```python
import mlflow

#Replace with your run id from mlflow localhost website
logged_model = 'runs:/f97aba50a14d4736963e7ee61050ec1c/model'

# Load model as a PyFuncModel.
loaded_model = mlflow.pyfunc.load_model(logged_model)

# Predict on a Pandas DataFrame.
import pandas as pd
predictions_df = loaded_model.predict(pd.DataFrame(data))
merged_data = pd.concat([data, predictions_df], axis=1)
print(merged_data)
```
