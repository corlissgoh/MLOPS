# Web Scraper for animeultima

### Importing packages

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Note: site being scraped could be a phishing site
```

### Get links for all animes under each category

```python
# URLs for different anime genres
tags = ["action", "adventure", "comedy", "dementia",'drama']
genres = ["Action", "Adventure", "Comedy", "Dementia", "Drama"]

# Base link
link = "https://aniwatchtv.to/genre/{}"
# https://aniwatchtv.to/genre/dementia
# Making a GET request
urls = {}
for tag in tags:
    formated_url = link.format(tag) 
    urls[genres[tags.index(tag)]] = formated_url

urls
```

```python
base_url = "https://aniwatchtv.to/genre"
movie_url = "https://aniwatchtv.to"
anime_links = {}

# Function to scrape pages (1-15) of each genre
def scrape_page(url,num):
    url = url + "?page=" + str(num)
    response = requests.get(url)
    html = response.content
    soup = bs(html, "html.parser")
    
    # Getting all the links for the anime
    hrefs = soup.find_all('a', class_="dynamic-name")
    for href in hrefs:
        print(href)
        link = movie_url + href.get('href')
        anime_links[genre].append(link)

animes =  {}

for genre in urls.keys():
    anime_links[genre] = []
    num = 1
    
    while num <= 15: 
    # As there are a lot of animes per genre, we will only scrape the first 15 pages of each genre
        url = urls[genre]
        print(url)
        response = requests.get(url)
        html = response.content
        soup = bs(html, "html.parser")
        scrape_page(url,num)
        num += 1
    anime = pd.DataFrame(anime_links[genre], columns=['link'])
    anime['genre'] = genre
    animes[genre] = anime
    print(f"{genre}: {len(anime_links[genre])}")

animes.keys()
```

```python
anime['link'][0]
```

```python
anime = pd.concat(animes.values(), ignore_index=True)
anime
```

```python
anime['link']= anime['link'].str.replace('/genre','')
anime
```

```python
# Remove duplicates
anime = anime.drop_duplicates(subset=['link'],ignore_index=True,keep='first')
anime
```

```python
titles = []
descriptions = []

for link in anime.loc[:,'link']:
    try:
        print(link)
        response = requests.get(link)
        response.raise_for_status()  # Check if the request was successful

        html = response.content
        soup = bs(html, "html.parser")

        # Getting the title of the anime
        title = soup.select("h2.film-name")[0].get_text().strip()
#         print(title)

        titles.append(title)

        # Getting the description
        description = soup.select("div.film-description")[0].get_text().strip()
#         print(description)
        descriptions.append(description)
    except (requests.RequestException, IndexError) as e:
        # Handle potential exceptions (e.g., network errors, index errors)
        print(f"Error processing link {link}: {e}")

# Add the new columns to the existing DataFrame
anime['anime'] = ''
anime['description']=''
for i in range(len(titles)):
    anime['anime'][i] = titles[i]
    anime['description'][i] = descriptions[i]

# Display the DataFrame
anime
```
