# Exploratory Data Analysis

I started by importing libraries which allows me to read the data from my local computer to have a brief view of the dataset in general

```python
import pandas as pd
```

```python
data = pd.read_excel('01_residential_data.xlsx')
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# General information about the dataset
data.info()
```

There are 14 columns in this dataset with no null count values found

```python
# Check for duplicates
duplicates = data.duplicated()
print(duplicates)
```

No duplicate data can be found in this dataset, meaning it is clean to be used 

I will now go into exploration of the data to better understand it so as to utilize it for my predictions

```python
# Calculate remaining years of lease
current_year = pd.to_datetime('today').year
data['lease_commence_date'] = pd.to_datetime(data['lease_commence_date'], errors='coerce')
data['remaining_years'] = data['lease_commence_date'].apply(lambda x: current_year - x.year if pd.notnull(x) else None)
```

One of the features that is used when doing the prediction is remaining year of lease, hence I choose to train the model with the duration (reming years of lease)instead of just the commence date

```python
data = data.drop(columns=['postal_code','latitude', 'longitude','lease_commence_date','month'])
```

 Latitude, longtitude, month and postal code are not required to train the model as there is no visible relationship between them and the model, hence they are dropped

I then created visuals to have a breif view of what my data contains

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

numerical_columns = ['floor_area_sqm', 'cbd_dist', 'min_dist_mrt']


for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data[column], y=data['resale_price'])
    plt.title(f'{column} vs Resale Price')
    plt.xlabel(column)
    plt.ylabel('Resale Price')
    plt.show()
```

I plotted scatter plots for numerical columns such as min_dist_mrt, cbd_dist and flooe_square_metre against resale price, from the scatter plots, it is visivble that resale price is affected by these features though there are outliers present

```python
categorical_columns = ['town', 'flat_type', 'storey_range', 'flat_model']


for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=data[column], y=data['resale_price'])
    plt.title(f'Resale Price Distribution by {column}')
    plt.xlabel(column)
    plt.ylabel('Resale Price')
    plt.xticks(rotation=45)
    plt.show()
```

I used box plots for categorical features such as town, flat_type, storey range flat model to see their relationship with resale price. As seen in the graph, these features also effect resale price.

```python
correlation_matrix = data[numerical_columns].corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

I also plotted a correlation matrix for the numerical columns

# Modeling

This is the modeling part where the model will be created using pycaret and then saved into a pipeline for later use

```python
import mlflow

mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
pip install --upgrade pycaret mlflow
```

```python
data.head()
```

```python
# initialize setup
from pycaret.regression import *
s = setup(data, target = 'resale_price', 
          normalize=True,
          feature_selection=True,
          session_id=123,
          log_experiment = True, 
          fold_strategy='kfold',
          experiment_name = 'sale_hdb')
```

```python
best_model = compare_models(exclude=['rf', 'et'])
```

Random Forest Regression abdextratereeshasbeenremovedfromthisdatasethas been ommited as it is a complex model which consumes a lot of memory and also requires a lot of processing power. This means that the downloaded pipeline will be too huge to upload into github to be deployed. After some further exploration ,

```python
# check the residuals of trained model
plot_model(best_model, plot = 'residuals')
```

```python
# check feature importance
plot_model(best_model, plot = 'feature')
```

```python
# create a copy of data and drop resale_Price
data2 = data.copy()

data2.drop('resale_price', axis=1, inplace=True)
data2
```

```python
# generate predictions
from pycaret.regression import predict_model
predictions = predict_model(best_model, data=data2)
predictions.head()
```
