```python
# Convert 'genre','anime' and 'description' column to lowercase
df['genre'] = df['genre'].str.lower()
df['anime'] = df['anime'].str.lower()
df['description'] = df['description'].str.lower()

# Display the updated DataFrame
df
```

```python
# remove stopwords
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK stopwords data
nltk.download('stopwords')
nltk.download('punkt')


# Function to remove stopwords
def remove_stopwords(text):
    if isinstance(text, str):
        stop_words = set(stopwords.words('english'))
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
        return ' '.join(filtered_text)
    else:
        return text

# Apply the function to the 'description' column
df['anime'] = df['anime'].apply(remove_stopwords)
df['description'] = df['description'].apply(remove_stopwords)

# Display the updated DataFrame
df

```

```python
# remove punctuation
```

```python
import pandas as pd
import string



# Function to remove punctuation
def remove_punctuation(text):
    if isinstance(text, str):
        punctuation_chars = string.punctuation
        cleaned_text = ''.join(char for char in text if char not in punctuation_chars)
        return cleaned_text
    else:
        return str(text)  # Convert non-string (e.g., NaN) to string

# Apply the function to the 'anime' and 'description' column
df['anime'] = df['anime'].apply(remove_punctuation)
df['description'] = df['description'].apply(remove_punctuation)

# Display the updated DataFrame
df
```

```python
# remove short words & digits
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK stopwords data
nltk.download('stopwords')
nltk.download('punkt')

# Function to remove short words, digits, and stopwords
def remove_stopwords(text):
    if isinstance(text, str):
        stop_words = set(stopwords.words('english'))
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if len(word) > 2 and not word.isdigit() and word.lower() not in stop_words]
        return ' '.join(filtered_text)
    else:
        return str(text)

# Apply the function to the 'anime' and description' column
df['anime'] = df['anime'].apply(remove_stopwords)
df['description'] = df['description'].apply(remove_stopwords)

# Display the updated DataFrame
df
```

```python
# lemmitization, filter out lemmitizations and keep nouns
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')


# Function for lemmatization and filtering nouns
def lemmatize_and_filter_nouns(text):
    if isinstance(text, str):
        stop_words = set(stopwords.words('english'))
        lemmatizer = WordNetLemmatizer()

        word_tokens = word_tokenize(text)
        tagged_words = pos_tag(word_tokens)

        # Lemmatize and filter out stopwords and non-nouns
        lemmatized_text = [lemmatizer.lemmatize(word, pos='n') for word, pos in tagged_words if len(word) > 2 and not word.isdigit() and word.lower() not in stop_words and pos.startswith('N')]
        
        return ' '.join(lemmatized_text)
    else:
        return str(text)

# Apply the function to the 'anime' and description' column
df['anime'] = df['anime'].apply(lemmatize_and_filter_nouns)
df['description'] = df['description'].apply(lemmatize_and_filter_nouns)

# Display the updated DataFrame
df
```
