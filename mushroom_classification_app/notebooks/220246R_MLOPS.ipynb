### Step 4: Visualize the Data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=df)
plt.title('Distribution of Mushroom Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

The bar chart illustrates the distribution of mushroom classes in the dataset, with two categories: edible and poisonous. The chart shows a near-equal distribution of mushrooms between the two categories, with a slight edge for edible mushrooms. This balanced distribution indicates that the dataset provides a sufficient amount of data for both classes, enabling the ML model to learn effectively from both edible and poisonous mushrooms.

The visualization highlights the importance of considering both classes when training the ML model, as it has a balanced representation of each. This balanced representation can help the model avoid biases and improve its ability to generalize, ultimately leading to more accurate predictions for both edible and poisonous mushrooms.

```python
# Count plots for categorical variables
categorical_columns = df.columns[1:]  # Exclude the 'class' column

plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(6, 4, i)
    sns.countplot(x=column, data=df, order=df[column].value_counts().index)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This is a visualization of the distribution of different mushroom features, showing the frequency of each value within each feature.

- `gill-spacing`: Close gill spacing is highly prevalent, with crowded spacing being less common.

- `stalk-root`: Bulbous stalks dominate the dataset, indicating a bulbous base.
- `stalk-surface-above-ring`: Smooth stalk surfaces above the ring are the most prevalent, suggesting a smooth texture.
- `stalk-surface-below-ring`: Smooth stalk surfaces below the ring are the most frequent, suggesting a smooth texture.
- `stalk-color-above-ring`: White stalk color above the ring dominates, suggesting a common color pattern.
- `stalk-color-below-ring`: White stalk color below the ring dominates, suggesting a common color pattern.

Edible Mushrooms:

- There are also some edible mushrooms with red, pink, buff, and purple caps.

Therefore, we can conclude that the color of the cap is a potentially useful feature for distinguishing between poisonous and edible mushrooms. However, it is important to note that the color of the cap is not a foolproof indicator, and there are some poisonous mushrooms that have the same cap color as edible mushrooms.

```python
# Lift Chart
plot_model(best_model, plot='lift')
```

The lift chart shows how well the Decision Tree Classifier can identify the most relevant cases compared to a random selection. The x-axis represents the percentage of the sample considered, and the y-axis represents the lift, or the ratio of the number of true positives found using the model to the number of true positives found by random chance. The higher the lift, the better the model's performance.

In this case, the lift curve for both classes (0 and 1) starts high and then gradually decreases, indicating that the model is effective in identifying the most relevant cases early on. However, as the percentage of the sample increases, the lift decreases, suggesting that the model's performance diminishes for less relevant cases.

### Step 2: Create a Streamlit Script

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pycaret.classification as pc

# Load the model
model = pc.load_model('mushroom-pipeline')

# Streamlit UI
st.title('Mushroom Species Prediction')
st.write("Enter the details to predict if the mushroom is edible or poisonous")

# Collect input data from the user
cap_shape = st.selectbox('Cap Shape', ['convex', 'bell', 'conical', 'flat', 'knobbed', 'sunken'])
cap_surface = st.selectbox('Cap Surface', ['smooth', 'scaly', 'fibrous', 'grooves'])
cap_color = st.selectbox('Cap Color', ['brown', 'buff', 'cinnamon', 'gray', 'green', 'pink', 'purple', 'red', 'white', 'yellow'])
bruises = st.selectbox('Bruises', ['bruises', 'no'])
odor = st.selectbox('Odor', ['almond', 'anise', 'creosote', 'fishy', 'foul', 'musty', 'none', 'pungent', 'spicy'])
gill_attachment = st.selectbox('Gill Attachment', ['attached', 'free'])
gill_spacing = st.selectbox('Gill Spacing', ['close', 'crowded', 'distant'])
gill_size = st.selectbox('Gill Size', ['broad', 'narrow'])
gill_color = st.selectbox('Gill Color', ['black', 'brown', 'buff', 'chocolate', 'gray', 'green', 'orange', 'pink', 'purple', 'red', 'white', 'yellow'])
stalk_shape = st.selectbox('Stalk Shape', ['enlarging', 'tapering'])
stalk_root = st.selectbox('Stalk Root', ['bulbous', 'club', 'equal', 'rhizomorph', 'rooted', 'missing'])
stalk_surface_above_ring = st.selectbox('Stalk Surface Above Ring', ['fibrous', 'scaly', 'smooth'])
stalk_surface_below_ring = st.selectbox('Stalk Surface Below Ring', ['fibrous', 'scaly', 'smooth'])
stalk_color_above_ring = st.selectbox('Stalk Color Above Ring', ['brown', 'buff', 'cinnamon', 'gray', 'orange', 'pink', 'red', 'white', 'yellow'])
stalk_color_below_ring = st.selectbox('Stalk Color Below Ring', ['brown', 'buff', 'cinnamon', 'gray', 'orange', 'pink', 'red', 'white', 'yellow'])
veil_type = st.selectbox('Veil Type', ['partial', 'universal'])
veil_color = st.selectbox('Veil Color', ['brown', 'orange', 'white', 'yellow'])
ring_number = st.selectbox('Ring Number', ['none', 'one', 'two'])
ring_type = st.selectbox('Ring Type', ['cobwebby', 'evanescent', 'flaring', 'large', 'none', 'pendant', 'sheathing'])
spore_print_color = st.selectbox('Spore Print Color', ['black', 'brown', 'buff', 'chocolate', 'green', 'orange', 'pink', 'purple', 'red', 'white', 'yellow'])
population = st.selectbox('Population', ['abundant', 'clustered', 'numerous', 'scattered', 'several', 'solitary'])
habitat = st.selectbox('Habitat', ['grasses', 'leaves', 'meadows', 'paths', 'urban', 'waste', 'woods'])

# Create a DataFrame from user input
input_data = pd.DataFrame({
    'cap-shape': [cap_shape],
    'cap-surface': [cap_surface],
    'cap-color': [cap_color],
    'bruises': [bruises],
    'odor': [odor],
    'gill-attachment': [gill_attachment],
    'gill-spacing': [gill_spacing],
    'gill-size': [gill_size],
    'gill-color': [gill_color],
    'stalk-shape': [stalk_shape],
    'stalk-root': [stalk_root],
    'stalk-surface-above-ring': [stalk_surface_above_ring],
    'stalk-surface-below-ring': [stalk_surface_below_ring],
    'stalk-color-above-ring': [stalk_color_above_ring],
    'stalk-color-below-ring': [stalk_color_below_ring],
    'veil-type': [veil_type],
    'veil-color': [veil_color],
    'ring-number': [ring_number],
    'ring-type': [ring_type],
    'spore-print-color': [spore_print_color],
    'population': [population],
    'habitat': [habitat]
})

# Make predictions
prediction = model.predict(input_data)
st.write(f'The predicted class is: {prediction[0]}')
```
